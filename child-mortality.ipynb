## Modelling 
**We can start modelling our data ,where my target variable is a continuous one**

### Linear Regression

**Now we will apply Linear regression upon the selected columns in order to regression analysis**


```python
def nan_filler(dt,col):
    li = np.unique(dt[pd.isna(dt[col])]["Country"])
    dt = dt[~dt["Country"].isin(li)]
    return dt
    
```

```python
data = nan_filler(data,"Prevalence of anemia among pregnant women (%)")
data = nan_filler(data,"DTP3 (% of one-year-olds immunized)")
data = nan_filler(data,"GDP per capita")
```

```python
linear_data = data[['Maternal_mortality_ratio','Prevalence of anemia among pregnant women (%)','DTP3 (% of one-year-olds immunized)','GDP per capita','Mortality rate_under_5_per_1000']]
x = linear_data.drop(["Mortality rate_under_5_per_1000"],axis=1)
y = linear_data["Mortality rate_under_5_per_1000"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 0,test_size=0.25)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
y_pred = regr.predict(X_train)
```

```python
print("R squared: {}".format(r2_score(y_true=y_train,y_pred=y_pred)))
```

**Here we see R squared = 0.899 which means our model is doing good**

```python
residuals = y_train.values-y_pred
mean_residuals = np.mean(residuals)
print("Mean of Residuals {}".format(mean_residuals))
```

**The Mean of Residuals are close to 0 , which means the model is performing good**

### Homoscedasticity
**Homoscedasticity means that the residuals have equal or almost equal variance across the regression line. By plotting the error terms with predicted terms we can check that there should not be any pattern in the error terms**


<br>**Now we will be apply Goldfeld Quandt Test**</br>
**Null Hypothesis: Error terms are homoscedastic**</br>
**Alternative Hypothesis: Error terms are heteroscedastic**</br>

```python
p = sns.scatterplot(y_pred,residuals)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot([0,26],[0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
```

**Since p value is more than 0.05 in Goldfeld Quandt Test, we can't reject it's null hypothesis that error terms are homoscedastic.Which signifies our Model is performing good**



```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip
name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(residuals, X_train)
lzip(name, test)
```

```python
p = sns.distplot(residuals,kde=True)
p = plt.title('Normality of error terms/residuals')
```

 **It is clear that residuals are normally distributed**<br>
 **Hence the assumptions for Linear Regression are all met here**
